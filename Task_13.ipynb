Task 1: Student Result Processing System
Create a class Student with:

Private attributes: __marks

Methods:

add_marks()

calculate_grade()

display_result()

Logic

Grade A: ≥ 80

Grade B: ≥ 60

Grade C: ≥ 40

Fail: < 40

```python
class student:
    def __init__(self):
        self.__marks = 0

    def add_marks(self, m):
        self.__marks = m

    def calculate_grade(self):
        if self.__marks >= 80:
            return "Grade A"
        elif self.__marks >= 60:
            return "Grade B"
        elif self.__marks >= 40:
            return "Grade C"
        else:
            return "Fail"

    def display_result(self):
        print("marks:", self.__marks)
        print("result:", self.calculate_grade())


s = student()
s.add_marks(76)
s.display_result()
```

Task 2: Product Inventory Management
Create a class Product with:

Product ID, name, price, stock

Method to:

Update stock

Apply discount

Display product info

```python
class Product:
    def __init__(self, pid, name, price, stock):
        self.pid = pid
        self.name = name
        self.price = price
        self.stock = stock

    def update_stock(self, qty):
        self.stock += qty

    def apply_discount(self, percent):
        self.price -= self.price * percent / 100

    def display(self):
        print(self.pid, self.name, self.price, self.stock)
p = Product(1, "Phone", 20000, 10)
p.update_stock(5)
p.apply_discount(10)
p.display()
```

Task 3: Secure Bank Account System
Create a class BankAccount:

Private variables: __balance, __pin

Methods:

deposit()

withdraw() (validate PIN)

check_balance()

```python
class BankAccount:
    def __init__(self, balance, pin):
        self.__balance = balance
        self.__pin = pin

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt, pin):
        if pin == self.__pin:
            self.__balance -= amt
        else:
            print("Wrong PIN")

    def check_balance(self):
        print(self.__balance)
acc = BankAccount(5000, 1234)
acc.deposit(1000)
acc.withdraw(500, 1111)
acc.check_balance()
```

Task 4: Employee Salary Management
Create a class Employee:

Private salary

Setter with validation (no negative salary)

Method to calculate annual salary

```python
class Employee:
    def __init__(self):
        self.__salary = 0

    def set_salary(self, sal):
        if sal > 0:
            self.__salary = sal
        else:
            print("Invalid salary")

    def annual_salary(self):
        return self.__salary * 12
e = Employee()
e.set_salary(30000)
print(e.annual_salary())
```

Task 8: Payment Gateway System
Create abstract class Payment

Method: pay(amount)

Implement:

UPI

Card

Wallet

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class UPI(Payment):
    def pay(self, amount):
        print("UPI payment:", amount)

class Card(Payment):
    def pay(self, amount):
        print("Card payment:", amount)

class Wallet(Payment):
    def pay(self, amount):
        print("Wallet payment:", amount)

p = [UPI(), Card(), Wallet()]
for i in p:
    i.pay(500)
```

Task 9: Notification System
Abstract class Notification

Method: send()

Implement:

EmailNotification

SMSNotification

```python
from typing import Self
from abc import ABC, abstractmethod
class Notification(ABC):
  @abstractmethod
  def send(self):
    pass

class EmailNotification(Notification):
  def send(self):
    print("email send")

class SMSNotification(Notification):
  def send(self):
    print("SMS send")

e= EmailNotification()
e.send()

s= SMSNotification()
s.send()
```

Task 10: Shape Area Calculator Base class Shape • Method area() Child classes: • Circle • Rectangle • Triangle

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        print(3.14 * 5 * 5)

class Rectangle(Shape):
    def area(self):
        print(4 * 6)

class Triangle(Shape):
    def area(self):
        print(0.5 * 4 * 6)
```
