```python
#Task 1: Bank Account System (Single Inheritance)
class BankAccount:
    def __init__(self, acc_no, balance):
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amt):
        self.balance += amt

    def withdraw(self, amt):
        self.balance -= amt

    def display(self):
        print(self.acc_no, self.balance)


class SavingsAccount(BankAccount):
    def __init__(self, acc_no, balance, interest):
        super().__init__(acc_no, balance)
        self.interest = interest

    def display(self):
        print(self.acc_no, self.balance, self.interest)


s = SavingsAccount(101, 5000, 5)
s.deposit(1000)
s.withdraw(500)
s.display()
```

```python
#Task 2: Shape Area Calculator (Multilevel Inheritance)
class Shape:
    def area(self):
        pass


class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        print(self.l * self.w)


class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)


r = Rectangle(4, 5)
r.area()

s = Square(4)
s.area()
```

```python
#Task 3: Library Management System (Hierarchical Inheritance)
class LibraryItem:
    def __init__(self, title, year):
        self.title = title
        self.year = year

    def display(self):
        print(self.title, self.year)


class Book(LibraryItem):
    def __init__(self, title, year, author):
        super().__init__(title, year)
        self.author = author

    def display(self):
        print(self.title, self.year, self.author)


class Magazine(LibraryItem):
    def __init__(self, title, year, issue):
        super().__init__(title, year)
        self.issue = issue

    def display(self):
        print(self.title, self.year, self.issue)


b = Book("Python", 2023, "Guido")
m = Magazine("Tech", 2024, 12)

b.display()
m.display()
```

```python
#Task 4: Online Shopping Cart (Multiple Inheritance)
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price


class Discount:
    def apply_discount(self, percent):
        return self.price - (self.price * percent / 100)


class CartItem(Product, Discount):
    def __init__(self, name, price):
        Product.__init__(self, name, price)


c = CartItem("Laptop", 50000)
print(c.apply_discount(10))
```

```python
#Task 5: School Staff and Students (Constructor + Method Inheritance)
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Teacher(Person):
    def __init__(self, name, age, subject, salary):
        super().__init__(name, age)
        self.subject = subject
        self.salary = salary

    def display(self):
        print(self.name, self.age, self.subject, self.salary)


class Student(Person):
    def __init__(self, name, age, grade, sid):
        super().__init__(name, age)
        self.grade = grade
        self.sid = sid

    def display(self):
        print(self.name, self.age, self.grade, self.sid)


t = Teacher("Ravi", 35, "Maths", 40000)
s = Student("Amit", 16, "10th", 101)

t.display()
s.display()
```
