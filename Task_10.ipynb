```python
#Task 11: Simple Interest (default rate = 5%)
def simple_interest(p, t, r=5):
    return (p * t * r) / 100

print(simple_interest(1000, 2))
```

```python
#task:16 Sum of any number of numbers
def add_numbers(*args):
    total = 0
    for n in args:
        total = total + n
    return total
add_numbers(11,22,56)
```

```python
#task:17 Write a function that prints all arguments passed to it.
def show_args(*args):
    for item in args:
        print(item)
show_args("cow",1,5.7,"wool")
```

```python
#task:18 Function that MULTIPLIES all numbers
def multiply_numbers(*args):
    result = 1
    for n in args:
        result = result * n
    return result
multiply_numbers(11,22,56)
```

```python
#task:19 Create a function that prints all key–value pairs passed to it.
def show_items(**kwargs):
    for key in kwargs:
        print(key, kwargs[key])
show_items(name="mani", age=20, city="Delhi")
```

```python
#task:20 Write a function create_profile(**kwargs) that stores any user details.
def create_profile(**kwargs):
    return kwargs
profile= create_profile(name="rat",sub="computer")
print(profile)
```

```python
#task:21 Function to calculate the total price from items passed as item=price.
def total_price(**items):
    total = 0
    for item in items:
        total = total + items[item]
    return total
total_price(apple=30, milk=20)
```

```python
#task:22 Write a recursive function to calculate factorial.
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)
factorial(5)
```

```python
#task:24 Write a recursive function to print numbers from 1 to n.
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)

print_numbers(5)
```

```python
#task:27 Sort a list of tuples using a lambda key
data = [(1, 3), (4, 1), (2, 5)]
data.sort(key=lambda x: x[1])
print(data)
```
