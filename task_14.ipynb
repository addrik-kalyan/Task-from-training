Print Numbers and Letters

Thread 1: Print numbers 1–10

Thread 2: Print letters A–J

```python
import threading
class A:
    def letter(self):
        print("A,B,C,D,E,F,G")

class B(A):
    def number(self, i):
        print(i)

    def start_threads(self):
        for i in range(1,11):
            t = threading.Thread(target=self.number, args=(i,))
            t.start()

obj = B()
obj.letter()
obj.start_threads()
```

Download Simulation

Thread 1: Print “Downloading file 1…”

Thread 2: Print “Downloading file 2…”

Use time.sleep() to simulate delay



```python
import time

class A:
    def down(self):
      print("Downloading file 1…")
      time.sleep(2)


class B(A):
    def load(self):
      print("Downloading file 2…")
      time.sleep(2)

obj = B()
obj.down()
obj.load()
```

Task:3

Shared Counter

Two threads increment a shared counter 5 times each

Use Lock() to avoid conflicts


```python
import threading

counter = 0
lock = threading.Lock()

def increment():
    global counter
    for i in range(5):
        with lock:
            counter += 1
            print(counter)

t1 = threading.Thread(target=increment)
t2 = threading.Thread(target=increment)

t1.start()
t2.start()
```

Task:4

Even-Odd Checker

Thread 1: Print even numbers 1–20

Thread 2: Print odd numbers 1–20

```python
import threading
class A:
  def even_numbers(self):
      for i in range(1,21):
        if i % 2 == 0:
          print("even number:.",i)

class B(A):
  def odd_numbers(self):
      for i in range(1,21):
        if i % 2 == 1:
          print("odd numbers:",i)
obj = B()
obj.even_numbers()
obj.odd_numbers()
```

task:5

Message Printer

Create 3 threads

Each prints a different message every second

Run for 5 iterations

```python
import threading
import time

def msg(message):
    for i in range(5):
        print(message)
        time.sleep(1)

t1 = threading.Thread(target=msg, args=("Hello",))
t2 = threading.Thread(target=msg, args=("Welcome",))
t3 = threading.Thread(target=msg, args=("Bye",))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()
```
