1 D array

```python
import numpy as np

# 1
arr = np.arange(1, 21)

# 2
print(arr[0])

# 3
print(arr[-1])

# 4
print(arr[7])

# 5
arr[4] = 100

# 6
print(arr[:6])

# 7
print(arr[5:13])

# 8
print(arr[-5:])

# 9
print(arr[::2])

# 10
print(arr[::-1])
```

2D array

```python
# 11
arr2 = np.arange(1, 13).reshape(3, 4)

# 12
print(arr2[1, 2])

# 13
print(arr2[0])

# 14
print(arr2[:, -1])

# 15
print(arr2[:2])

# 16
print(arr2[:, -2:])

# 17
print(arr2[1, 1:3])

# 18
print(arr2[:2, :2])

# 19
arr2[arr2 == 6] = 60

# 20
arr2[2] = 0
```

3D array

```python
# 21
arr3 = np.arange(1, 25).reshape(2, 3, 4)

# 22
print(arr3[0])

# 23
print(arr3[0, 1, 2])

# 24
print(arr3[1])

# 25
print(arr3[0, :, :])

# 26
print(arr3[:, 0, :])

# 27
print(arr3[:, :, -1])

# 28
arr3[0, 0, 0] = 999

# 29
arr3[arr3 > 15] = 0

# 30
print(arr3[:, :2, :2])
```
