```python
#4. Multiplication table using a function
def table(num):
    for i in range(1, 11):
        print(num, "x", i, "=", num * i)

table(5)
```

```python
#5. Function to return squares from 1 to n
def squares(n):
    for i in range(1, n + 1):
        print(i, "square is", i*i)

squares(5)
```

```python
#6. Function to check even or odd
def even_or_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(7))
```

```python
#7. Function to calculate factorial
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(factorial(5))
```

```python
#8. Function to find sum of digits
def sum_of_digits(num):
    total = 0
    for digit in str(num):   # convert number to string
        total += int(digit)
    return total

print(sum_of_digits(123))
```

```python
#9. Prime check using for-else
def is_prime(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, num):
        if num % i == 0:
            break
    else:
        return "Prime"

    return "Not Prime"

print(is_prime(6))
```

```python
#10. Search element in list using for-else
numbers = [1, 3, 5, 7, 9]
search = 7

for i in numbers:
    if i == search:
        print("Found!")
        break
else:
    print("Not Found!")
```

```python
#11. Break when a vowel is found
text = "sky"

for ch in text:
    if ch in "aeiouAEIOU":
        print("Vowel found:", ch)
        break
else:
    print("No vowel found")
```

```python
#13. Sum from 1 to 100 using range()
total = 0

for i in range(1, 101):
    total += i

print("Sum is:", total)
```

```python
#14. Function to reverse a string
def reverse_string(text):
    return text[::-1]

print(reverse_string("hello"))
```

```python
#15. Check if all numbers are positive (for-else)
numbers = [2, 4, -1, 6]

for num in numbers:
    if num < 0:
        print("Negative number found:", num)
        break
else:
    print("All numbers are positive")
```
