```python
#Task 14: Count Frequency Using Set + Loop
words = ["hi", "hello", "hi", "hello", "welcome"]

unique_words = set(words)
freq = {}

for word in unique_words:
    count = 0
    for w in words:
        if w == word:
            count += 1
    freq[word] = count

print(freq)
```
