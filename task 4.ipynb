```python
#Task 2: Find the frequency of each element
n = [1, 2, 2, 3, 3, 3]

print("1 appears", n.count(1), "times")
print("2 appears", n.count(2), "times")
print("3 appears", n.count(3), "times")
```

```python
#Task 3: Find the common elements between two lists
a = [1,2,3,4]
b = [3,4,5,6]

common = []

if 3 in a and 3 in b:
    common.append(3)
if 4 in a and 4 in b:
    common.append(4)

print(common)
```
