```
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))
op = input("Enter the operator (+, -, *, /, %): ")

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b if b != 0 else "Cannot divide by zero")
elif op == "%":
    print(a % b)
else:
    print("Invalid operator")
```
