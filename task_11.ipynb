```python
#Task 1: Default Constructor
class Laptop:
    def __init__(self):
        self.brand = "HP"
        self.ram = "8GB"

    def display(self):
        print(self.brand, self.ram)

l1 = Laptop()
l1.display()
```

```python
#Task 2: Parameterized Constructor
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display(self):
        print(self.name, self.roll_no)

s1 = Student("Amit", 1)
s2 = Student("Ravi", 2)

s1.display()
s2.display()
```

```python
#Task 3: Class with Destructor
class FileHandler:
    def __init__(self):
        print("File opened")

    def __del__(self):
        print("File closed")

f = FileHandler()
del f
```

```python
#Task 4: Constructor + Destructor Together
class Employee:
    def __init__(self):
        print("Employee created")

    def __del__(self):
        print("Employee removed")

def fun():
    e = Employee()

fun()
```

```python
#Task 5: Count Objects Using Constructor & Destructor
class Counter:
    count = 0

    def __init__(self):
        Counter.count += 1
        print("Count:", Counter.count)

    def __del__(self):
        Counter.count -= 1
        print("Count:", Counter.count)

a = Counter()
b = Counter()
del a
del b
```

```python
#Task 6: Real-World Simulation
class DatabaseConnection:
    def __init__(self):
        print("Connected to database")

    def __del__(self):
        print("Connection closed")

db = DatabaseConnection()
del db
```
