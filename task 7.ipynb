```python
#Task:1 Even or Odd
num = int(input("Enter number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#Task:2 Positive, Negative, or Zero
num = float(input("Enter number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#Task:3 Largest of Three Numbers
a = int(input("Enter first: "))
b = int(input("Enter second: "))
c = int(input("Enter third: "))

if a >= b and a >= c:
    print("Largest =", a)
elif b >= a and b >= c:
    print("Largest =", b)
else:
    print("Largest =", c)
```

```python
#task:4 Leap Year Checker
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#task:5 Vowel or Consonant
ch = input("Enter a letter: ").lower()

if ch in ["a","e","i","o","u"]:
    print("Vowel")
else:
    print("Consonant")
```

```python
#task:6 Simple Calculator
a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))
op = input("Enter operator (+ - * /): ")

if op == "+":
    print("Result =", a + b)
elif op == "-":
    print("Result =", a - b)
elif op == "*":
    print("Result =", a * b)
elif op == "/":
    print("Result =", a / b)
else:
    print("Invalid operator")
```

```python
#task:7 Print 1 to 100
for i in range(1, 101):
    print(i)
```

```python
#task:8 Sum of First N Natural Numbers
n = int(input("Enter N: "))
total = 0

for i in range(1, n + 1):
    total = total + i

print("Sum =", total)
```

```python
#task:9 Multiplication Table
n = int(input("Enter number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

```python
#task:10 Reverse a Number
n = int(input("Enter number: "))
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print("Reversed =", rev)
```

```python
#task:11 Count Digits
n = int(input("Enter number: "))
count = 0

while n > 0:
    count = count + 1
    n = n // 10

print("Total digits =", count)
```

```python
#task:12 Factorial
n = int(input("Enter number: "))
fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial =", fact)
```
