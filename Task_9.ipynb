```python
#1. Print all even numbers between 50 and 150
for i in range(50, 151):
    if i % 2 == 0:
        print(i)
```

```python
#2. Print numbers from 100 to 10 decreasing by 5
for i in range(100, 9, -5):
    print(i)
```

```python
#3. Generate a list of cubes from 1 to 30
cubes = []

for num in range(1, 31):
    cubes.append(num * num * num)
print(cubes)
```

```python
#4. Print numbers divisible by both 4 and 6 (1 to 200)
for num in range(1, 201):
    if num % 4 == 0 and num % 6 == 0:
        print(num)
```

```python
#5. Multiplication tables (1–10) for 2, 3, and 4
for num in [2, 3, 4]:
    print("Table of", num)

    for i in range(1, 11):
        print(num, "x", i, "=", num * i)
```

```python
#6. Print elements at odd positions (index-based
my_list = [10, 20, 30, 40, 50, 60]

for i in range(1, len(my_list), 2):
    print(my_list[i])
```

```python
#7. Count numbers between 1 and 500 ending with digit 7
count = 0

for num in range(1, 501):
    if num % 10 == 7:
        count += 1

print("Total:", count)
```

```python
#8. Print the pattern
for num in range(1, 6):
    print(str(num) * num)
```

```python
#9. List of three-digit numbers with strictly increasing digits
numbers = []

for num in range(100, 1000):
    a = num // 100
    b = (num // 10) % 10
    c = num % 10

    if a < b and b < c:
        numbers.append(num)

print(numbers)
```

```python
#10. Print all hour–minute combinations (24-hour format, every 10 minutes)
for hour in range(0, 24):
    for minute in range(0, 60, 10):
        print(hour, ":", minute)
```
